```python
from numpy.random import randint
import pandas as pd
import matplotlib.pyplot as plt

def roland_wright(x:int)->dict:
    """
    param: x
    roll die x times, record how many rolls were required before a six resulted in a dictionary
    """
    
    # init dict to store results
    results = dict()
    
    # perform x die rolls
    # reset counter each iteration
    for _ in range(x):
        count = 0
        res = 0
        
        # roll until 6, keep track of count
        while res != 6:
            count += 1
            res = randint(1, 7)
        
        # store result in dict
        if count in results.keys():
            results[count] +=1
        else:
            results[count] = 1
        
        
    return(results)
```

```python
die_throws = pd.Series(roland_wright(10000))
```

```python
import seaborn as sns
```

```python
die_throws_df = (die_throws
                 .reset_index()
                 .rename(columns={'index':'Rolls before 6', 0: 'n'})
                 .sort_values('n', ascending=False)
                )
```

```python
fig, ax = plt.subplots(figsize=(14, 6))

ax = sns.barplot(
    y='Rolls before 6', x='n', data=die_throws_df.query('n > 20'),
    orient='h', color='midnightblue', alpha=0.8
)

ax.set_title('x = 10000', loc="left", size=16)

plt.savefig('rolling_dice_on_a_saturday.png')
```
