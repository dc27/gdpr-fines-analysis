```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
df = pd.read_excel("https://github.com/chris1610/pbpython/blob/master/data/sample-salesv3.xlsx?raw=true")
```

```python
df.info()
```

```python
top10 = (
    df.groupby('name')
    .agg({'ext price': 'sum', 'quantity' : 'count'})
    .sort_values(by='ext price', ascending=False)
    .head(10)
    .reset_index()
    .rename(columns={'name': 'Name', 'ext price': 'Sales', 'quantity': 'Purchases'})
)
```

```python
top10
```

```python
plt.style.use('ggplot')
```

```python
top10.plot(kind='barh', y='Sales', x='Name')
```

```python
fig, ax = plt.subplots()
top10.plot(kind='barh', y='Sales', x='Name', color= 'steelblue', ax=ax)
```

```python
fig, ax = plt.subplots()
top10.plot(kind='barh', y='Sales', x='Name', ax=ax)
ax.set_xlim([-10000, 140000])
ax.set_xlabel('\nTotal Revenue')
ax.set_ylabel(None);
```

```python
fig, ax = plt.subplots()
top10.plot(kind='barh', y="Sales", x="Name", ax=ax)
ax.set_xlim([-10000, 140000])
ax.set(title='2014 Revenue', xlabel='Total Revenue', ylabel=None);
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
top10.plot(kind='barh', y="Sales", x="Name", ax=ax)
ax.set_xlim([-10000, 140000])
ax.set(title='2014 Revenue', xlabel="Total Revenue")
ax.legend().set_visible(False)
```

```python
def currency(x, pos):
    """
    The two args are the value and tick position
    """
    if x >= 100000:
        return '${:1.1f}M'.format(x*1e-6)
    return '${:1.0f}K'.format(x*1e-3)
```

```python
fig, ax = plt.subplots()
top10.plot(kind='barh', y="Sales", x="Name", ax=ax)
ax.set_xlim()
ax.set(title='2014 Revenue', xlabel='Total Revenue', ylabel=None)
formatter = FuncFormatter(currency)
ax.xaxis.set_major_formatter(formatter)
ax.legend().set_visible(False)
```

```python
fig, ax = plt.subplots()

top10.plot(kind='barh', y='Sales', x='Name', ax=ax)
avg = top10.Sales.mean()

ax.set_xlim([-10000, 140000])
ax.set(title="2014 Revenue", xlabel="\nTotal Revenue", ylabel="Customer")

ax.axvline(x=avg, color='b', label="Average", linestyle="--")


for cust in [3, 5, 8]:
    ax.text(115000, cust, "New Customer")
    
formatter = FuncFormatter(currency)
ax.xaxis.set_major_formatter(formatter)

ax.legend().set_visible(False)
```
