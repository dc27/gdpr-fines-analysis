```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

richmondway = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2023/2023-09-26/richmondway.csv')
```

```python
richmondway.columns
```

```python
sns.catplot(x="Episode_order", y="F_count_RK", data=richmondway, col="Season", sharex=False, hue="Coaching_flag")
```

```python
sns.regplot(x="F_count_RK", y="Imdb_rating", data=richmondway, color="Season")
```

```python
sns.barplot(x="Season", y="F_count_RK", data=richmondway.groupby("Season").F_count_RK.mean().reset_index(), color="steelblue")
```
